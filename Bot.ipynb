#Chat Bot

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import json
import pickle
import random
import numpy as np
```

```python
words=[]
classes=[]
documents=[]
ignore_words=['?','!','.']
data_file = open('/content/intents.json').read()
intents = json.loads(data_file)
```

```python
intents['intents'][0]
```

```python
for intent in intents['intents']:
	for pattern in intent['patterns']:
		#tokenize here
		w=nltk.word_tokenize(pattern)
		#print('Token is: {}'.format(w))
		words.extend(w)
		#(['hey', 'you'], 'greeting')
		documents.append((w, intent['tag']))
		# add the tag to classes list
		if intent['tag'] not in classes:
			classes.append(intent['tag'])
	
```

```python
	# Final lists
	print('Words list is: {}'.format(words))
	print('Docs are: {}'.format(documents))
	print('Classes are: {}'.format(classes))
```

```python
words = [lemmatizer.lemmatize(w.lower()) for w in words if w not in ignore_words]
words = list(set(words))
classes = list(set(classes))
print('words {}'.format(words))
print('classes {}'.format(classes))
```

```python
pickle.dump(words, open('words.pkl', 'wb'))
pickle.dump(classes, open('classes.pkl', 'wb'))
```

```python
training = []
output_empty = [0]*len(classes)                 #
# [0,0,0,0,0,0,0,0]
for doc in documents:
	bag = []
	pattern_words = doc[0]
	pattern_words = [lemmatizer.lemmatize(word.lower()) for word in pattern_words]
	#print('Current Pattern Words: {}'.format(pattern_words))

	for w in words:
		bag.append(1) if w in pattern_words else bag.append(0)

	#print('Current Bag: {}'.format(bag))

	output_row = list(output_empty)
	output_row[classes.index(doc[1])] = 1
	#print('Current Output: {}'.format(output_row))

	training.append([bag, output_row])
```

```python
print('Current Pattern Words: {}'.format(pattern_words))  #for the last output
print('Current Bag: {}'.format(bag))                      #for the last output
print('Current Output: {}'.format(output_row))            #for the last output
print('Training: {}'.format(training))
```

```python
random.shuffle(training)                   #So as to prevent formation of sequence thus preventing overfitting
training = np.array(training)
```

```python
train_x = list(training[:, 0])          #bag
train_y = list(training[:, 1])          #output_row
print('X: {}'.format(train_x))
print('Y: {}'.format(train_y))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))

# compiling the model & define an optimizer function
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',optimizer=sgd, metrics=['accuracy'])
```

```python
mfit = model.fit(np.array(train_x), np.array(train_y), epochs=200, batch_size=5, verbose=1)
```

```python
model.save('chatbot_model.h5', mfit)
```

```python
from keras.models import load_model
```

```python
intents = json.loads(open('/content/intents.json').read())
model = load_model('chatbot_model.h5')
words = pickle.load(open('words.pkl', 'rb'))
classes = pickle.load(open('classes.pkl', 'rb'))
```

```python
def bow(sentence):
	sentence_words = nltk.word_tokenize(sentence)
	sentence_words = [lemmatizer.lemmatize(word.lower()) for word in sentence_words]
	bag = [0]*len(words)
	for s in sentence_words:
		for i, w in enumerate(words):
			if w == s:
				bag[i]=1
	return (np.array(bag))
```

```python
def predict_class(sentence):
	sentence_bag = bow(sentence)
	res = model.predict(np.array([sentence_bag]))[0]
	ERROR_THRESHOLD = 0.25
	results = [[i,r] for i,r in enumerate(res) if r > ERROR_THRESHOLD]
	#sort by probablity
	results.sort(key=lambda x: x[1], reverse=True)       #So as to get the highest probability value i.e. the r value
	return_list = []
	for r in results:
		return_list.append({'intent':classes[r[0]], 'probablity':str(r[1])})
	return return_list 
```

```python
def getResponse(ints):
	tag = ints[0]['intent']
	list_of_intents = intents['intents']
	for i in list_of_intents:
		if(i['tag']==tag):
			result=random.choice(i['responses'])
			break

	return result
```

```python
def chatbot_response(msg):
	ints = predict_class(msg)
	res = getResponse(ints)
	return res
```
